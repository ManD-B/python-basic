# 서울시 CCTV 현황 데이터 분석

```python
import numpy as np
import pandas as pd
import csv
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
mpl.rc('font', family='Malgun Gothic')
%matplotlib inline
```

## 1. 서울통계서비스 사이트에서 '주민등록인구(구별)' 통계에서 2018년도 데이터를 다운 받아서 '서울시 인구현황.txt' 파일로 저장한다.

## 2. 파일에서 데이터를 읽어 컬럼이 '자치구', '인구수', '내국인', '외국인', '고령자' 인 데이터 프레임을 만든다.(pop)

```python
pop = pd.read_csv('서울시 인구현황.txt', sep = '\t', thousands = ',', usecols = [1, 3, 6, 9, 13], skiprows = 4, names = ['자치구', '인구수', '내국인', '외국인', '고령자'])
pop
```

## 3. 외국인 비율과 고령자 비율을 추가한다.

```python
pop['외국인 비율'] = pop['외국인'] / pop['인구수'] * 100
pop['고령자 비율'] = pop['고령자'] / pop['인구수'] * 100
pop.head()
```

## 4. 서울시 자치구 년도별 CCTV 설치 현황 데이터를 다운 받아서 '서울시 자치구 년도별 CCTV 설치 현황.xlsx' 파일로 저장한후, 데이터를 읽서 데이터 프레임을 만든다.(cctv)

```python
cctv = pd.read_excel('서울시 자치구 년도별 CCTV 설치 현황.xlsx')
cctv
```

## 5. '기관명' 컬럼 데이터의 빈칸을 없애고, 컬럼명을 '자치구'로 변경한다.

```python
cctv['기관명'].unique()
```

```python
# cctv['기관명'] = ['강남구', '강동구', '강북구', '강서구', '관악구', '광진구', '구로구',
#                '금천구', '노원구', '도봉구', '동대문구', '동작구', '마포구', '서대문구',
#                '서초구', '성동구', '성북구', '송파구', '양천구', '영등포구', '용산구',
#                '은평구', '종로구', '중구', '중랑구']
cctv['기관명'] = [x.replace(' ', '') for x in cctv['기관명'].unique()]
cctv.head()
```

```python
cctv.rename(columns = {'기관명' : '자치구'}, inplace = True)
cctv.head()
```

## 6. 2015년 이전의 데이터를 합하여 '2015년 이전' 컬럼을 만들고, 최근 3개년간의 데이터를 합하여 '최근 3개년' 컬럼을 만든 후, '최근 증가율' 컬럼을 만든다.

```python
cctv[['2011년 이전', '2012년', '2013년']] = \
    cctv[['2011년 이전', '2012년', '2013년']].fillna(0).astype(int)
cctv['2015년 이전'] = cctv[['2011년 이전', '2012년', '2013년', '2014년', '2015년']].sum(axis = 1)
cctv['최근 3개년'] = cctv.iloc[:, [7, 8, 9]].sum(axis = 1)
cctv['최근 증가율'] = cctv['최근 3개년'] / cctv['2015년 이전'] * 100
cctv.head()
```

## 7. '자치구', '소계', '최근 증가율' 컬럼을 제외하고 나머지는 모두 삭제한다.

```python
cctv.drop(['2011년 이전', '2015년 이전', '최근 3개년'], axis = 1, inplace = True)
del cctv['2012년']
del cctv['2013년']
del cctv['2014년']
del cctv['2015년']
del cctv['2016년']
del cctv['2017년']
del cctv['2018년']
cctv.head()
```

## 8. 서울시 인구 데이터와 cctv 데이터를 합치고, '자치구'를 인덱스로 만든다.

```python
df = pd.merge(cctv, pop, on = '자치구')
df.head()
```

```python
df.set_index('자치구', inplace = True)
df.head()
```

## 9. CCTV 현황을 다양한 형태로 시각화하여 보여준다.

```python
plt.rcParams['figure.figsize'] = (14, 12)
plt.rcParams['font.size'] = 14
```

## 자치구별 CCTV 댓수 (내림차순으로)

```python
plt.title('자치구별 CCTV 댓수')
df['소계'].sort_values().plot(kind = 'barh', grid = True)
plt.show()
```

## 자치구별 인당 CCTV 댓수

```python
df['인당 CCTV'] = df['소계'] / df['인구수']
df['인당 CCTV'].sort_values().plot(kind = 'barh', grid = True)
plt.title('자치구별 인당 CCTV 댓수')
plt.show()
```

## - 자치구별 최근증가율(내림차순으로)

```python
df['최근 증가율'].sort_values().plot(kind = 'barh', grid = True)
plt.title('자치구별 최근증가율')
plt.show()
```

## -인구수와 CCTV 댓수의 산점도 및 회귀선

```python
plt.scatter(df['인구수'], df['소계'])
plt.xlabel('인구수')
plt.ylabel('CCTV 대수')
plt.grid()
plt.show()
```

```python
fp1 = np.polyfit(df['인구수'], df['소계'], 1)
fp1
```

```python
f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 101)
```

```python
plt.figure(figsize = (12, 10))
plt.scatter(df['인구수'], df['소계'])
plt.plot(fx, f1(fx), ls = 'dashed', lw = 3, color = 'r')
plt.xlabel('인구수')
plt.ylabel('CCTV 대수')
plt.grid()
plt.show()
```

## -인구수와 CCTV 댓수의 산점도, 회귀선 및 자치구명

```python
df.head()
```

```python
df['오차'] = np.abs(df['소계'] - f1(df['인구수']))
df.head()
```

```python
df_sort = df.sort_values(by = '오차', ascending = False)
df_sort.head(10)
```

```python
plt.figure(figsize = (14, 10))
plt.scatter(df['인구수'], df['소계'], c = df['오차'], s = 50)
plt.plot(fx, f1(fx), ls = 'dashed', lw = 1, color = 'g')

for n in range(10) :
    plt.text(df_sort['인구수'][n] * 1.02, df_sort['소계'][n] * 0.98,
             df_sort.index[n], fontsize = 15) # 텍스트 1.02는 옆으로, 0.98은 높이를 나타냄
plt.xlabel('인구수')
plt.ylabel('인구당비율')
plt.colorbar()
plt.grid()
plt.show()
```
