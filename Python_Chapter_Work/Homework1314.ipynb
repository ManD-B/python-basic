```python
import pandas as pd
import csv
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
mpl.rc('font', family='Malgun Gothic')
%matplotlib inline
```

```python
# 3.
df = pd.read_csv('high_school_2019.csv')
sum = df[' 1학년 학생수'] + df[' 2학년 학생수'] + df[' 3학년 학생수']
df['총 학생수'] = sum
df
```

```python
# 4.
frame = df.iloc[:, [3, 5, 7]]
sum = frame.sum(axis = 0)
print('1학년 총 학생 수 : %d' %sum[0])
print('2학년 총 학생 수 : %d' %sum[1])
print('3학년 총 학생 수 : %d' %sum[2])
plt.bar(['1학년', '2학년', '3학년'], [sum[0], sum[1], sum[2]])
plt.title('전국 고등학교 학년별 총 학생수')

plt.show()                      
```
