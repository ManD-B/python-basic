```python
# Matplotlib 사용, 엑셀처리 및 한글 처리를 위한 Code
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import csv
%matplotlib inline
mpl.rcParams['axes.unicode_minus'] = False # 한국어 사용
mpl.rc('font', family='Malgun Gothic') # 폰트 변경
plt.rcParams["figure.figsize"]= (15,9) # 그래프 크기 변경
```

```python
# 1.
f = open('cafe_2year.csv', 'r', encoding = 'utf-8')
lines = csv.reader(f)

header = next(lines)

x = []
y = []
for line in lines:
    if '12' in line[0] or '03' in line[0] or '06' in line[0] or '09' in line[0]:
        x.append(line[0])
        y.append(line[1])

plt.bar(x, y)
plt.title('전국의 월별 카페수', size = 18, color = 'green')
plt.xlabel('년월')
plt.ylabel('카페수')
plt.show()
```

```python
# 2.
f = open('doctor_2019.csv', 'r', encoding='utf-8')
lines = csv.reader(f)

header = next(lines)

reg = ['서울', '부산', '대구', '인천', '대전', '광주', '울산']

data = np.zeros((7,4), dtype = np.int32)

for line in lines :
    if line[0] == '서울':
        data[0][0] += int(line[2])
        data[0][1] += int(line[3])
        data[0][2] += int(line[4])
        data[0][3] += int(line[5])
    elif line[0] == '부산' :
        data[1][0] += int(line[2])
        data[1][1] += int(line[3])
        data[1][2] += int(line[4])
        data[1][3] += int(line[5])
    elif line[0] == '대구' :
        data[2][0] += int(line[2])
        data[2][1] += int(line[3])
        data[2][2] += int(line[4])
        data[2][3] += int(line[5])
    elif line[0] == '인천' :
        data[3][0] += int(line[2])
        data[3][1] += int(line[3])
        data[3][2] += int(line[4])
        data[3][3] += int(line[5])
    elif line[0] == '대전' :
        data[4][0] += int(line[2])
        data[4][1] += int(line[3])
        data[4][2] += int(line[4])
        data[4][3] += int(line[5])
    elif line[0] == '광주' :
        data[5][0] += int(line[2])
        data[5][1] += int(line[3])
        data[5][2] += int(line[4])
        data[5][3] += int(line[5])
    elif line[0] == '울산' :
        data[6][0] += int(line[2])
        data[6][1] += int(line[3])
        data[6][2] += int(line[4])
        data[6][3] += int(line[5])
print('지역  일반의  인턴  레지던트 전문의')
for i in range(len(reg)) :
    print(reg[i], end = '')
    for j in range(4) :
        print(' %6d' %data[i][j], end = '')
    print()
f.close()
```

```python
f2 = open('doctor2.csv', 'w', encoding='utf-8', newline ='')
wr = csv.writer(f2)

wr.writerow(['지역', '일반의 수', '인턴 수', '레지던트 수', '전문의 수'])
for i in range(len(reg)) :
    wr.writerow([reg[i], data[i][0], data[i][1], data[i][2], data[i][3]])
    
print('doctor2.csv 파일 쓰기 완료!')
f2.close()
```

```python
f2 = open('doctor2.csv', 'r', encoding='utf-8')
lines = csv.reader(f2)

header = next(lines)

x = []
y = []

for line in lines :
    x.append(line[0])
    y.append(int(line[4]))
    
plt.scatter(x, y)

plt.title('서울 및 광역시 종합병원 전문의 수', size =18 , color = 'green')
plt.xlabel('시')
plt.ylabel('의사 수')

plt.show()
f2.close()
```

```python
f2 = open('doctor2.csv', 'r', encoding='utf-8')
lines = csv.reader(f2)

header = next(lines)

a = []
b = []
c = []
d = []
e = []
for line in lines :
    a.append(line[0])
    b.append(int(line[1]))
    c.append(int(line[2]))
    d.append(int(line[3]))
    e.append(int(line[4]))
plt.plot(a, b, 'ro-', label='일반의', linewidth =2, markersize=7)
plt.plot(a, c, 'bx--',label='인턴', linewidth =2, markersize=7)
plt.plot(a, d, 'ys:', label='레지던트',linewidth =2, markersize=7)
plt.plot(a, e, 'md-.', label='전문의',linewidth =2, markersize=7)

plt.title('서울 및 광역시 종합병원 의사 수', size =18 , color = 'green')
plt.xlabel('시')
plt.ylabel('의사 수')

plt.legend(loc = 'best')
plt.show()
f2.close()
```

```python
f = open('doctor2.csv', 'r', encoding='utf-8')
lines = csv.reader(f)

header = next(lines)

kind = ['일반의', '인턴', '레지던트', '전문의']

numbers = []
seoul = next(lines)

total = 0
for i in range(4) :
    total += int(seoul[i+1])
    
for i in range(4) :
    persent = int(seoul[i+1])/total;
    numbers.append(persent)

plt.pie(numbers, explode=(0, 0.1, 0, 0), labels=kind, autopct='%.1f%%', shadow=True, startangle=90)
plt.title('서울 종합병원 의사 유형별 분포', size =18 , color = 'green')
plt.show()
        
f.close()
```
