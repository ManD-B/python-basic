```python
# 3.
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
mpl.rcParams['axes.unicode_minus'] = False # 한국어 사용
mpl.rc('font', family='Malgun Gothic') # 폰트 변경
plt.rcParams["figure.figsize"]= (15,9) # 그래프 크기 변경

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
 
ax.set_xlim([-7, 7])
ax.set_ylim([-1.2, 1.2])
ax.set_xticks([-np.pi * 2, -np.pi * 3 / 2, -np.pi, -np.pi/2, 0, np.pi/2, np.pi, np.pi * 3 / 2, np.pi * 2])
ax.set_xticklabels(['$-2*\pi$','$-3/2\pi$','$-\pi$', '$-\pi/2$', '$0$', '$\pi/2$', '$\pi$','$3/2\pi$','$2\pi$'])
 
x = np.linspace(-np.pi * 2, np.pi * 2, 101)
tanhy = np.tanh(x)
 
ax.plot(x, tanhy, color='blue', linewidth=2)
 
plt.show()
```

```python
from bs4 import BeautifulSoup 
from urllib.request import urlopen
import pandas as pd
import requests

header = {'User-Agent': 'Mozilla/5.0 (X11; CrOS i686 2268.111.0) AppleWebKit/536.11 (KHTML, like Gecko) Chrome/20.0.1132.57'}
 
url = "https://www.melon.com/chart/month/index.htm#params%5Bidx%5D=1&params%5BrankMonth%5D=202005&params%5BisFirstDate%5D=false&params%5BisLastDate%5D=true"
req = requests.get(url, headers=header)
html = req.text
soup = BeautifulSoup(html, "html.parser")

title = soup.find_all("div", {"class": "ellipsis rank01"})
artist = soup.find_all("span", {"class": "checkEllipsis"})
album = soup.find_all("div", {"class": "ellipsis rank03"})
rank = soup.find_all("span", {"class": "rank_wrap"})

title = [a.text.replace('\n', '') for a in title]
artist = [a.text for a in artist]
album = [a.text.replace('\n', '') for a in album]
rank = [a.text.replace('\n', '') for a in rank]
a = []
for i in range(1,101):
    a.append(i)
df = pd.DataFrame({'순위':a, '제목':title, '가수':artist, '앨범명':album, '전월순위':rank})
df.set_index('순위', inplace = True)
df.head(10)
```
